### Гипотеза №1: Если мы увеличим цену для поездок классом Комфорт и Бизнес, то выручка увеличится, потому что наблюдаются более низкие юзерские отмены и более длинные поездки.

```python
HW = pd.read_csv('Netology_A_B-testing_HW_1.csv')
HW
```

```python
# разбиваю юзеров на группы с помощью рандома
HW['value'] = np.random.random(101500)
HW.head()
```

```python
# в зависимости от значения value помещаю юзера в нруппу 1 или 2
HW['col1'] = HW['value'].apply(lambda x: 1 if x >= 0.5 else 2)
HW['col2'] = HW['value'].apply(lambda x: 1 if x < 0.5 else 2)
HW.head(7)
```

```python
# только те юзеры которые используют поездки классом Комфорт и Бизнес
order_class = HW[HW['order_class'].isin(['comfort','business'])]
order_class
```

```python
order_class.distance.fillna(order_class.distance.median() ,inplace=True) # в столбце distance  пропуски заменяем медианой 
order_class
```

```python
order_class = order_class.drop(index=[101498]) # уравниваем значение длины выборок
```

```python
order_class.info()
```

```python
group1 = order_class[order_class['col1']==1].distance
group1.head()
```

```python
group2 = order_class[order_class['col1']==2].distance
group2.head()
```

```python
# проверка однородности с помощью критерия Манна-Уитни
from scipy.stats import mannwhitneyu

stat, p = mannwhitneyu(group1,group2) # U-тест Манна-Уитни
print('Statistics=%.3f, p=%.3f' % (stat, p))

alpha = 0.05   # Уровень значимости
if p > alpha:
    print('Одинаковые распределения (не отвергаем H0)')
else:
    print('Разные распределения (отвергаем H0)')
```

### Гипотеза №2:Если увеличить цену за поездку в период Surge , то увидим увеличение выручки, потому что цена за поездку увеличится.

```python
# выбираю поездки в период Surge
surge = HW[HW['surge'].isin(['surge'])]
surge.head()
```

```python
surge = surge.drop(index=[21]) # уравниваем значение длины выборок
```

```python
surge.info()
```

```python
group3 = surge[surge['col1']==1].order_made
group3.head()
```

```python
group4 = surge[surge['col1']==2].order_made
group4.head()
```

```python
# проверка однородности с помощью критерия Манна-Уитни
from scipy.stats import mannwhitneyu

stat, p = mannwhitneyu(group3,group4) # U-тест Манна-Уитни
print('Statistics=%.3f, p=%.3f' % (stat, p))

alpha = 0.05   # Уровень значимости
if p > alpha:
    print('Одинаковые распределения (не отвергаем H0)')
else:
    print('Разные распределения (отвергаем H0)')
```
